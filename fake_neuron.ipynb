```python
from micrograd.engine import Value
from micrograd.viz import draw_graph

x1 = Value(2, 'x1')
x2 = Value(0, 'x2')
w1 = Value(-3, 'w1')
w2 = Value(1,'w2')
b = Value(6.8813735870195432, 'b')

x1w1 = x1 * w1
x2w2 = x2 * w2
added = x1w1 + x2w2
with_bias = added + b
o = with_bias.tanh()
```

```python
draw_graph(o)
```

```python
o.run_full_backpropagation()
```
